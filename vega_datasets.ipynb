```python
df=pd.read_csv('data/pokemon_Prasert.csv')
df.info()
```

```python
df.head(3)
```

```python
alt.Chart(df).mark_circle().encode(
    alt.X('attack'),
    alt.Y('defense'),
    alt.Size('speed'),
    alt.Color('class'),
    alt.Tooltip(['hp','percent-male','percent-female']),
    alt.Order('speed', sort='descending'),
    alt.Row('type1'),
    alt.Column('type2'),
)
```
